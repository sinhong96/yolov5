```python

img_folder_path='D:\\hong\\ReSearch\\object_detection\\project11\\fooddata\\images\\train\\01_012_01012005_161198475166549_0.jpeg'
img=cv2.imread(img_folder_path)
dh, dw, _ = img.shape

ac_bbox = 'D:\\hong\\ReSearch\\object_detection\\project11\\fooddata\\labels\\train1\\01_012_01012005_161198475166549_0.txt'
bbox_filename = ac_bbox

with open(bbox_filename , "r") as f:
    boxes_coors = f.readlines()
    
    for dt in boxes_coors:
        # Split string to float
        _, x, y, w, h = map(float, dt.strip().split(' ')) # x_center, y_center, width, height
        
        l = int((x - w / 2) * dw) # left
        r = int((x + w / 2) * dw) # right
        t = int((y - h / 2) * dh) # top
        b = int((y + h / 2) * dh) # bottom
        
        if l < 0:
            l = 0 # prevent -ve value
        if r > dw - 1:
            r = dw - 1 # set max width of bbox as the width of img 
        if t < 0:
            t = 0 # prevent -ve value
        if b > dh - 1:
            b = dh - 1 # set max height of bbox as the height of img 
        
        # cv2.rectangle(image, start_point, end_point, color_of_box, Line_thickness)
        cv2.rectangle(img, (l, t), (r, b), (0, 0, 255), 3)
        
plt.figure(figsize=(18,18))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
    
```
